```python
import damask
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
res = damask.Result('DP_linear_UAT_RD_tensionX.hdf5')
```

```python
# List of sigma_vM and epsilon vM for separate phases
sigma_ferrite = []
sigma_martensite = []
for s in res.get('sigma_vM').values():
    sigma_ferrite.append( np.average(s['BCC_Ferrite']))
    sigma_martensite.append(np.average(s['BCC_Martensite']))
E_ferrite = []
E_martensite = []
for s in res.get('epsilon_V^0.0(F)_vM').values():
    E_ferrite.append(np.average(s['BCC_Ferrite']))
    E_martensite.append(np.average(s['BCC_Martensite']))
```

```python
# Plot ferrite and martensite stress strain curves
plt.figure()
plt.plot(E_ferrite, sigma_ferrite, label = 'Ferrite curve')
plt.plot(E_martensite, sigma_martensite, label = 'Martensite curve')
plt.legend()
plt.xlabel('strain')
plt.ylabel('stress')
plt.show()
```
